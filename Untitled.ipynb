```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('EV_Population.csv')
data.head()
```

```python
# Display the dataset's structure
print("Basic Dataset Information:")
print(data.info())
print("\nDataset Head:")
print(data.head())
```

```python
print("\nOriginal Column Names:")
print(data.columns.tolist())
```

```python
data.columns = data.columns.str.strip().str.lower().str.replace(" ", "_")
print("\nNormalized Column Names:")
print(data.columns.tolist())
```

```python
district_column = "legislative_district"  # Adjust based on normalized names
ev_count_column = "electric_range"  # Adjust based on normalized names
```

```python
if district_column not in data.columns or ev_count_column not in data.columns:
    raise ValueError(f"Columns '{district_column}' or '{ev_count_column}' not found in the dataset!")
else:
    print(f"Columns '{district_column}' and '{ev_count_column}' successfully found.")
```

```python
data = data.dropna(subset=[district_column, ev_count_column])
```

```python
data[ev_count_column] = pd.to_numeric(data[ev_count_column], errors="coerce")
data[district_column] = data[district_column].astype(str)
```

```python
print("\nSummary Statistics:")
print(data.describe())
```

```python
district_summary = data.groupby(district_column)[ev_count_column].sum().reset_index()
```

```python
district_summary = district_summary.sort_values(by=ev_count_column, ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=district_summary, x=district_column, y=ev_count_column, palette="viridis")
plt.title("Impact of Legislative Districts on EV Adoption")
plt.xlabel("Legislative District")
plt.ylabel("Total EV Range")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
```

```python
top_10_districts = district_summary.head(10)
plt.figure(figsize=(8, 8))
plt.pie(
    top_10_districts[ev_count_column],
    labels=top_10_districts[district_column],
    autopct="%1.1f%%",
    startangle=140,
    colors=sns.color_palette("pastel")
)
plt.title("Top 10 Districts by EV Adoption")
plt.tight_layout()
```

```python
numeric_columns = data.select_dtypes(include=[np.number]).columns
if len(numeric_columns) > 1:
    print("\nCorrelation Matrix:")
    correlation_matrix = data[numeric_columns].corr()
    print(correlation_matrix)
```

```python
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Correlation Matrix")
    plt.tight_layout()
```

```python
top_5_districts = district_summary.head(5)
data['is_top_5'] = data[district_column].isin(top_5_districts[district_column])
```

```python
grouped = data.groupby('is_top_5')[ev_count_column].mean()
print("\nAverage EV Range - Top 5 vs Other Districts:")
print(grouped)
```

```python
district_summary["cumulative_ev_range"] = district_summary[ev_count_column].cumsum()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=district_summary, x=district_column, y="cumulative_ev_range", marker="o", label="Cumulative Range")
plt.xticks(rotation=45, ha="right")
plt.title("Cumulative EV Adoption by Legislative District")
plt.xlabel("Legislative District")
plt.ylabel("Cumulative EV Range")
plt.legend()
plt.tight_layout()
```
